```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# read simulation outputs

filename = '../simulations/ngspice/tb_lskm_g1_433_20250919.raw'
with open(filename, 'r') as f:
    d1 = f.readlines()

variables = int(d1[5][15:])
print(variables)
points = int(d1[6][12:])
print(points)

var_names = []
for i in range(0,variables):
    var_names.append(d1[8+i])
```

```python
def read_data_from_index(points,variables,d,j):
    global freq,absz0,q0,absz1,q1
    for i in range(0,points):
        aux = d[j+(variables+1)*i]
        aux = aux.split()[1]
        freq.append(float(aux.split(',')[0]))
        aux = d[j+1+(variables+1)*i]
        absz0.append(float(aux.split(',')[0]))
        aux = d[j+2+(variables+1)*i]
        q0.append(float(aux.split(',')[0]))
        aux = d[j+3+(variables+1)*i]
        absz1.append(float(aux.split(',')[0]))
        aux = d[j+4+(variables+1)*i]
        q1.append(float(aux.split(',')[0]))


freq = []
absz0 = []
q0 = []
absz1 = []
q1 = []

j = 14
read_data_from_index(points,variables,d1,j)
```
