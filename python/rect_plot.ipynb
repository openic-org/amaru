```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# read simulation outputs

filename = '../simulations/ngspice/tb_rect_g1_433_20250919.raw'
with open(filename, 'r') as f:
    d1 = f.readlines()

variables = int(d1[5][15:]) +1
print(variables)
points = int(d1[6][12:])
print(points)

var_names = []
for i in range(0,variables):
    var_names.append(d1[8+i])
```

```python
def read_data_from_index(points,variables,d,j):
    global t,va,vb,vo1,vo2,vo3,vout,iv1,iv3,ix1m1,ix1m2,ix1m3,ix1m4,ix4m1,ix4m2,ix4m3,ix4m4
    for i in range(0,points):
        aux = d[j+(variables)*i]
        a = aux.split()[1]
        t.append(float(a))
        va.append(float(d[j+1+(variables)*i]))
        vb.append(float(d[j+2+(variables)*i]))
        vo1.append(float(d[j+3+(variables)*i]))
        vo2.append(float(d[j+4+(variables)*i]))
        vo3.append(float(d[j+5+(variables)*i]))
        vout.append(float(d[j+6+(variables)*i]))
        iv1.append(float(d[j+7+(variables)*i]))
        iv3.append(float(d[j+8+(variables)*i]))
        ix1m1.append(float(d[j+9+(variables)*i]))
        ix1m2.append(float(d[j+10+(variables)*i]))
        ix1m3.append(float(d[j+11+(variables)*i]))
        ix1m4.append(float(d[j+12+(variables)*i]))
        ix4m1.append(float(d[j+13+(variables)*i]))
        ix4m2.append(float(d[j+14+(variables)*i]))
        ix4m3.append(float(d[j+15+(variables)*i]))
        ix4m4.append(float(d[j+16+(variables)*i]))


t = []
va = []
vb = []
vo1 = []
vo2 = []
vo3 = []
vout = []
iv1 = []
iv3 = []
ix1m1 = []
ix1m2 = []
ix1m3 = []
ix1m4 = []
ix4m1 = []
ix4m2 = []
ix4m3 = []
ix4m4 = []

j = 26
read_data_from_index(points,variables,d1,j)
```
